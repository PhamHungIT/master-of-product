```python
# Nhận thấy có nhiều giá trị thể hiện hãng Lenovo, ta thực hiện sửa đổi dữ liệu

# Xóa các dấu cách thừa ở đầu và cuối
df['brand'] = df['brand'].str.strip()

# Chuyển các cụm 'Lenovo Flagship', 'Lenovo Thinkpad', 'Lenovo Ideapad' thành 'Lenovo'
df['brand'] = df['brand'].replace(['Lenovo Flagship', 'Lenovo Thinkpad', 'Lenovo Ideapad'], 'Lenovo')
```

```python
# Số lượng máy tính của mỗi hãng
groupby_brand = df.groupby(['brand'])['name'].count().sort_values()
groupby_brand

# vẽ biểu đồ thống kê số lượng máy của từng hãng

groupby_brand.plot(kind='barh', fontsize=12, color = color[0])
plt.xlabel('Số lượng', fontsize=12)
plt.ylabel('Hãng', fontsize=12)
plt.title('Số lượng máy của từng hãng', fontsize=12)
plt.show()
```

```python
# 10 hãng máy tính có số lượng máy lớn nhất
groupby_brand_result = df.groupby(['brand'])['name'].count().sort_values(ascending=False).head(10)
groupby_brand_result
```

```python
groupby_brand_series = df.groupby(['brand'])['name'].count()
```

```python
data = {
    'brand': groupby_brand_series.index.to_list(),
    'count': groupby_brand_series.values
}
groupby_brand_df = pd.DataFrame(data)
```

```python
groupby_brand_df.loc[groupby_brand_df['count'] < 10, 'brand'] = 'other'
```

```python
groupby_brand_series = groupby_brand_df.groupby('brand')['count'].sum()
```

```python
no_colors = groupby_brand_series.count() 
colors = sns.color_palette('pastel')[0:no_colors]

# Vẽ biểu đồ tròn
plt.pie(groupby_brand_series.values, labels = groupby_brand_series.index, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
# Vẽ biểu đồ thể hiện mức giá trung bình của các hãng
groupby_brand_mean = df.groupby('brand')['price'].mean().sort_values()
# Vẽ biểu đồ thống kê số lượng máy của các hãng
plt.xticks(rotation='vertical')
plt.xlabel('Hãng', fontsize=12)
plt.ylabel('Giá x 1000đ', fontsize=12)
sns.barplot(x = groupby_brand_mean.index, y = groupby_brand_mean.values/1000)
```

```python
brands = list(df['brand'].unique())
brands
```

```python
from matplotlib.pyplot import figure
data1 = []
for brand in brands:
    data1.append(df[df['brand'] == brand]['price'])

fig, ax = plt.subplots()
ax.ticklabel_format(style='plain')
ax.boxplot(data1)
ax.set_xticklabels(brands, fontsize = 7);
plt.title(" Biểu đồ boxplot giá bán của các hãng laptop")
figure(figsize=(10, 10))
```

```python
from matplotlib.pyplot import figure

for brand in brands:
    graph_name = '{} - Boxplot Laptop Price'.format(brand)
    data = df[df['brand'] == brand]['price']
    fig, ax = plt.subplots()
    # Thay đổi định dạng dữ liệu trên trục y
    ax.ticklabel_format(style='plain')
    sns.boxplot(y=data)
    plt.title(graph_name)
    figure(figsize=(8, 6))
    plt.show()
```
